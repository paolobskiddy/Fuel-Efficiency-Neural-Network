```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_selection import mutual_info_regression
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, InputLayer
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import MeanSquaredError
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, OneHotEncoder, OrdinalEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import root_mean_squared_error
from imblearn.over_sampling import RandomOverSampler
from category_encoders import MEstimateEncoder
import numpy as np
import seaborn as sns
import scipy
from sklearn.svm import SVR 
```

```python
df = pd.read_csv("car_data.csv")
df.head()
```

```python
df_clean = df[df['model']!='i-miev']
X = df_clean.copy().fillna(0)
y = X.pop("combination_mpg")

for col_name in X.select_dtypes("object"):
    X[col_name], _ = X[col_name].factorize()
discrete_feature = X.dtypes == int
```

```python
def mi_score_check(X, y, discrete_feature):
    mi_scores = mutual_info_regression(X, y, discrete_features=discrete_feature)
    mi_scores = pd.Series(mi_scores, name="MI SCORE", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores

mi_scores = mi_score_check(X, y, discrete_feature)
mi_scores.index
```

```python
enc_ohe = OneHotEncoder(sparse_output=False, handle_unknown="infrequent_if_exist")
scaler = StandardScaler()
sampler = RandomOverSampler()
def data_transform(train, test, dtype):
    if dtype == "ord":
        X_ord_train = scaler.fit_transform(train)
        X_ord_test = scaler.transform(test)
        return X_ord_train, X_ord_test
        
    if dtype == "cat":
        X_cat_train = enc_ohe.fit_transform(train)
        X_cat_test =  enc_ohe.transform(test)
        return X_cat_train, X_cat_test
```

```python
ordinal = ['city_mpg','cylinders','displacement','combination_mpg', 'highway_mpg']
ordinal_2 = ['make_model',"avg_mpg", "class"]
ordinal_corr = ['city_mpg','cylinders','displacement', 'combination_mpg',"avg_mpg", "highway_mpg", 'make_model', "class"]
categorical = ['drive', 'class']

X = df_clean.copy()
y = X.copy().pop("combination_mpg")
#feature_engineering
X_1 = pd.DataFrame()
X_1 = X['make'].astype(str) + " " + X["model"].astype(str) + " " +X["year"].astype(str)
X_data = X[ordinal+categorical]
X_data.loc[:,["make_model"]] = X_1 
X_data.loc[:,["L_per_cyl"]] = (X_data["cylinders"] / X_data["displacement"])
X_data.loc[:,["L_per_cyl"]] = X_data["L_per_cyl"].mul(-1)
X_data.loc[:,["avg_mpg"]] = (X_data["city_mpg"].mul(0.55) + X_data["highway_mpg"].mul(0.45))
```

```python
X_data_mi = X_data.copy().drop(["combination_mpg"], axis=1)
for col_name in X_data_mi.select_dtypes("object"):
    X_data_mi[col_name], _ = X_data_mi[col_name].factorize()
discrete_feature = X_data_mi.dtypes == int

mi_scores_new = mi_score_check(X_data_mi, y, discrete_feature)
mi_scores_new.index

fig, ax = plt.subplots()
plt.bar(mi_scores_new.index, mi_scores_new)
plt.xticks(rotation=90)
plt.show()
```

```python
def MAE_calc(target, predicted):
    y_target, y_pred = np.array(target), np.array(predicted)
    MAE = np.mean(np.abs(y_target - y_pred))/target.shape[0]
    return MAE
error = MAE_calc(X_data["combination_mpg"],X_data["avg_mpg"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y, test_size = 0.20, random_state = 42)
encoder = MEstimateEncoder(cols=["make_model", "class"], m=5.0)
X_train = encoder.fit_transform(X_train, y_train)
X_test =  encoder.transform(X_test)

#X_train = X_train.drop(["combination_mpg","cylinders","displacement", "city_mpg", "highway_mpg"], axis=1)
#X_test = X_test.drop(["combination_mpg","cylinders","displacement", "city_mpg", "highway_mpg"], axis=1)
X_train
```

```python
model_neural =  Sequential()
model_neural.add(InputLayer(shape=[X_train_complete.shape[1],]))
model_neural.add(Dense(128, activation=None))
model_neural.add(Dense(11, activation=None))
model_neural.add(Dense(1, activation=None))
model_neural.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mse', RootMeanSquaredError])
```

```python
model_neural.fit(X_train_complete, y_train, epochs=200, validation_split=0.2)
```

```python
pred_n = model_neural.predict(X_test_complete)
rmse_n = root_mean_squared_error(y_test, pred_n)
print("\nRMSE: " + str(rmse_n))
```
